```python
# import allthe libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import titanic data set
file_path = ('tested.csv')


#load the data set
titanic_data = pd.read_csv(file_path)


#show the first five rows of titanic data set
titanic_data.head()
```

```python

print("Shape of dataset:", titanic_data.shape)
```

```python
# information about column
titanic_data.info()
```

```python
titanic_data.isnull().sum()
```

```python
# Fill missing Age with the median age
titanic_data['Age'].fillna(titanic_data['Age'].median(), inplace=True)

# Fill missing Fare with the median fare
titanic_data['Fare'].fillna(titanic_data['Fare'].median(), inplace=True)

# Check again if missing values exist
print(titanic_data.isnull().sum())

titanic_data.head()
```

### SURVIVAL COUNT

```python
plt.figure(facecolor='lightgray')
sns.countplot(data = titanic_data, x = 'Survived')
plt.title('survival count')
plt.xlabel('Survived (1 = Yes, 0 = No)')
plt.ylabel('Number Of Passenger')
plt.show()
```

### SURVIVAL BY GENDER

```python
# Plot Survival by Gender
sns.countplot(data=titanic_data, x='Sex', hue='Survived')
plt.title('Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.legend(['Did Not Survive', 'Survived'])
plt.show()

```

### Survival by Passenger Class

```python

# Plot Survival by Passenger Class
sns.countplot(data= titanic_data, x='Pclass', hue='Survived')
plt.title('Survival by Passenger Class')
plt.xlabel('Ticket Class (1 = 1st, 2 = 2nd, 3 = 3rd)')
plt.ylabel('Number of Passengers')
plt.legend(['Did Not Survive', 'Survived'])
plt.show()
```

### AGE DISTRIBUTION

```python
# Plot Age Distribution
plt.figure(figsize=(10,5))
sns.histplot(data= titanic_data, x='Age', bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```
